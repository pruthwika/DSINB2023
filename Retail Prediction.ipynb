```python
from sklearn.cluster import KMeans
```

```python
X=df.iloc[:,[2,3]].values
print(X)
wcss=[]

for i in range(1,11):
    kmeans=KMeans(n_clusters=i,init='k_means++',random_state=0)
    kmeans.fit()
    wcss.append(kmeans.intertia_)
    
```

```python
sns.set()
plt.plot(range(1,11), wcss)
plt.title('elbow graph')
plt.xlabel('No. of clusters')
plt.ylabel('Wcss')
plt.show()
```

```python
kmeans=KMeans(n_clusters=5,init='k-means++',random_state=0)
```

```python
Y=KMeans.fit_predict(X)

print(Y)
```

```python
plt.figure(figsize=(8,8))
plt.scatter(X[Y==0,0],X[Y==0,1],s=50,c='green',label='cluster1')
plt.scatter(X[Y==1,0],X[Y==1,1],s=50,c='blue',label='cluster2')
plt.scatter(X[Y==2,0],X[Y==2,1],s=50,c='red',label='cluster3')
plt.scatter(X[Y==3,0],X[Y==3,1],s=50,c='yellow',label='cluster4')
plt.scatter(X[Y==4,0],X[Y==4,1],s=50,c='gray',label='cluster5')

plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=100, c='cyan', label='centroids')
plt.show()
```
